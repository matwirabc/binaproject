### 3.1 Bereinigung der Daten
Im Rahmen der Analyse wurde eine systematische Datenbereinigung der Excel-Datei Struktur.xlsx vorgenommen. Diese Datei enthält für jedes Jahr zwischen 2013 und 2023 ein separates Tabellenblatt mit Bevölkerungsdaten nach Kanton. Die ursprünglichen Daten enthielten zusätzliche Formatierungen, Überschriften, leere Zeilen sowie weitere Spalten, die für die Analyse nicht relevant waren.

Die Bereinigung erfolgte in folgenden Schritten:

Auswahl der relevanten Tabellenblätter:
    Es wurden ausschliesslich die Blätter der Jahre 2013 bis 2023 berücksichtigt. Frühere Jahre wurden ausgeschlossen, da sie nicht Teil des Untersuchungszeitraums sind.

Zeilenbereinigung:
    Innerhalb jedes Tabellenblattes wurden nur die Zeilen 5 bis 36 (Index 4 bis 35) übernommen, da diese die 26 Kantone sowie einige zusätzliche Regionen enthalten. Oberhalb dieser Zeilen befanden sich nur Titelzeilen und Metadaten.

Spaltenauswahl:
    Es wurden nur die ersten neun Spalten (A bis I) einbezogen. Diese enthalten:
    Region (Kanton)
    Gesamtbevölkerung
    Altersgruppen: 0–19, 20–64, 65+
    Geschlecht: Männlich, Weiblich
    Staatsangehörigkeit: Schweizer, Ausländer

Spaltenbeschriftungen vereinheitlicht:
    Die ursprünglichen, teils mehrzeiligen oder sprachlich gemischten Spaltennamen wurden durch einheitliche, verständliche Bezeichnungen ersetzt:

"Region", "Total", "0–19", "20–64", "65+", "Männlich", "Weiblich", "Schweizer", "Ausländer"

Speicherung der bereinigten Daten:
    Die bereinigten Tabellen wurden sowohl:

in einer neuen Excel-Datei (Struktur_bereinigt.xlsx) gespeichert – mit einem Tabellenblatt pro Jahr, als auch in einem Python-Wörterbuch (daten_bereinigt) zur Weiterverarbeitung in Pandas geladen.

```python
def bereinige_rohdaten(eingabe_datei: str, ausgabe_datei: str) -> dict:
    """
    Bereinigt alle Tabellenblätter ab 2013 in der Bevölkerungs-Excel:
    • behält Zeilen 5–36   (Excel-Index 4–35)
    • behält Spalten A–I   (DataFrame-Index 0–8)
    • vergibt einheitliche Spaltennamen
    • schreibt jede Jahres-Tabelle in eine neue, saubere Excel-Datei
    • liefert ein Dict {jahr: DataFrame}
    """
    # Excel öffnen und relevante Blätter (≥2013) ermitteln
    xls = pd.ExcelFile(eingabe_datei)
    relevante_blaetter = sorted(
        [s for s in xls.sheet_names if s.isdigit() and int(s) >= 2013],
        key=int
    )

    # Spalten A–I (Index 0–8) & neue Überschriften
    keep_cols = list(range(0, 9))
    new_cols = [
        "Region", "Total", "0–19", "20–64", "65+",
        "Männlich", "Weiblich", "Schweizer", "Ausländer"
    ]

    # neue Arbeitsmappe anlegen
    wb = Workbook()
    wb.remove(wb.active)      # Standard-Sheet löschen

    jahres_dfs = {}

    for sheet in relevante_blaetter:
        df = pd.read_excel(eingabe_datei, sheet_name=sheet)

        # Zeilen & Spalten filtern, Spalten umbenennen
        df_clean = df.iloc[4:36, keep_cols].copy()
        df_clean.columns = new_cols

        jahres_dfs[sheet] = df_clean.copy()  # für die Weiterverarbeitung

        # in neues Sheet der Ziel-Datei schreiben
        ws = wb.create_sheet(title=sheet)
        for r in dataframe_to_rows(df_clean, index=False, header=True):
            ws.append(r)

    # Ordner anlegen (falls nicht vorhanden) & Datei speichern
    os.makedirs(os.path.dirname(ausgabe_datei), exist_ok=True)
    wb.save(ausgabe_datei)
    print(f"✓ Bereinigte Datei gespeichert unter: {ausgabe_datei}")

    return jahres_dfs

# ------------------------------------------------------------
# 2) Funktion aufrufen
# ------------------------------------------------------------
eingabe_datei = "Rohdaten/Struktur.xlsx"
ausgabe_datei = "Output/Struktur_bereinigt.xlsx"

daten_bereinigt = bereinige_rohdaten(eingabe_datei, ausgabe_datei)

# ------------------------------------------------------------
# 3) Alle Jahres-DataFrames zu einem grossen Table zusammenführen
# ------------------------------------------------------------
alle_jahre = []
for jahr, df in daten_bereinigt.items():
    df = df.copy()
    df["Jahr"] = int(jahr)
    alle_jahre.append(df)

gesamt_df = pd.concat(alle_jahre, ignore_index=True)

# ------------------------------------------------------------
```

## 4. Zusammenführung der Jahresdaten

Nach der Bereinigung der Bevölkerungsdaten für die Jahre 2013 bis 2023 liegen die Informationen in Form einzelner DataFrames vor – jeweils eines pro Jahr. Um eine konsistente Analyse über den gesamten Zeitraum zu ermöglichen, wurden diese Jahres-DataFrames in einem gemeinsamen Datensatz zusammengeführt.

```python
# Liste für alle Jahresdaten
alle_jahre = []

# Jeden Jahres-DataFrame mit Jahr versehen & hinzufügen
for jahr, df in daten_bereinigt.items():
    df = df.copy()
    df["Jahr"] = int(jahr)
    alle_jahre.append(df)

# Alle DataFrames zusammenfügen
gesamt_df = pd.concat(alle_jahre, ignore_index=True)

# Vorschau
gesamt_df.head()
```

### 4.1 Erstellung einer Pivot-Tabelle
Auf Basis des zusammengeführten DataFrames wird eine Pivot-Tabelle erstellt, um den zeitlichen Verlauf der Gesamtbevölkerung pro Region (z. B. Kantone oder Schweiz gesamt) besser analysieren zu können:

    Zeilen: Kalenderjahre (2013–2023)

    Spalten: Regionen (z. B. Zürich, Waadt, Schweiz)

    Werte: Bevölkerungszahl (Spalte „Total“)

Damit entsteht eine kompakte, matrixartige Darstellung, bei der jede Zelle die Einwohnerzahl einer Region in einem bestimmten Jahr enthält.

Um sicherzustellen, dass alle Daten korrekt numerisch interpretiert werden, wird im Anschluss explizit eine Typkonvertierung vorgenommen (pd.to_numeric). Diese Massnahme verhindert Fehler bei statistischen Operationen oder Visualisierungen.

```python
# Pivot-Tabelle: Jahr = Zeilen, Region = Spalten, Werte = Total
pivot_df = gesamt_df.pivot_table(
    index="Jahr",
    columns="Region",
    values="Total",
    aggfunc="sum"  # optional: "first" statt "sum"
).sort_index()

# Sicherstellen, dass alle Werte numerisch sind
pivot_df = pivot_df.apply(pd.to_numeric, errors="coerce")

# Sicherheitsprüfung: Datentyp und Werte
print(pivot_df.dtypes)
print(pivot_df.head())

# Filter für Kantone (ohne Regionen wie "Schweiz", "Zentralschweiz" etc.)
echte_kantone = [
    "Aargau", "Appenzell A. Rh.", "Appenzell I. Rh.", "Basel-Landschaft", "Basel-Stadt",
    "Bern", "Freiburg", "Genf", "Glarus", "Graubünden", "Jura", "Luzern", "Neuenburg",
    "Nidwalden", "Obwalden", "Schaffhausen", "Schwyz", "Solothurn", "St. Gallen",
    "Tessin", "Thurgau", "Uri", "Waadt", "Wallis", "Zug", "Zürich"
]

# Schweiz noch hinzugefügt für den summierten Wert. 
pivot_df = pivot_df[[col for col in pivot_df.columns if col in echte_kantone or col == "Schweiz"]]
```

### 5.1 Datenanalyse wie hat sich die Bevölkerungszahl gesamtschweizerisch und pro Kanton von 2013 bis 2023 entwickelt?
Im ersten Analyseschritt wird die demografische Entwicklung der gesamten Schweiz über den Zeitraum 2013 bis 2023 betrachtet. Dafür dient die zusammengeführte Pivot-Tabelle (siehe Abschnitt 2.5), in der die Zeile "Schweiz" die aggregierten Einwohnerzahlen pro Jahr enthält.

Die Analyse zeigt, dass die Gesamtbevölkerung der Schweiz im Beobachtungszeitraum von rund 8.14 Millionen Einwohnern im Jahr 2013 auf etwa 8.89 Millionen im Jahr 2023 angestiegen ist. Dies entspricht einem Wachstum von knapp 9 Prozent über zehn Jahre. Auffällig ist dabei die Kontinuität dieses Wachstums: Jahr für Jahr nahm die Bevölkerung zu, ohne dass es zu nennenswerten Rückgängen oder Wachstumsabbrüchen gekommen wäre. Dies deutet auf eine insgesamt stabile demografische Entwicklung hin, die weder durch politische noch durch wirtschaftliche Turbulenzen in diesem Zeitraum grundlegend beeinträchtigt wurde.


```python
# Daten vorbereiten
gesamtbev_pro_jahr = pivot_df["Schweiz"]
wachstumsrate = gesamtbev_pro_jahr.pct_change() * 100  # in %

# Plot
fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(
    gesamtbev_pro_jahr.index,
    gesamtbev_pro_jahr.values,
    color="black",
    marker="o",
    linewidth=2
)

# Prozentwerte über die Punkte schreiben
for jahr in gesamtbev_pro_jahr.index[1:]:  # ab 2014, da 2013 kein Vergleich hat
    wert = gesamtbev_pro_jahr[jahr]
    prozent = wachstumsrate[jahr]
    ax.annotate(
        f"{prozent:.1f}%", 
        (jahr, wert), 
        textcoords="offset points", 
        xytext=(0, 8), 
        ha='center', 
        fontsize=9
    )

# Achsen und Formatierungen
ax.set_title("Gesamtbevölkerung Schweiz (2013–2023)", fontsize=14)
ax.set_ylabel("Einwohnerzahl", fontsize=12)
ax.set_xlabel("Jahr", fontsize=12)
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f"{int(x):,}".replace(",", "'")))
ax.set_xticks(gesamtbev_pro_jahr.index)
ax.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

## 5.2 Welche Kantone verzeichnen das stärkste bzw. schwächste Bevölkerungswachstum (2013–2023)?
Um die regionalen Unterschiede in der Bevölkerungsentwicklung besser zu verstehen, wurde für jeden Kanton das prozentuale Wachstum der ständigen Wohnbevölkerung im Zeitraum 2013 bis 2023 berechnet. Grundlage dafür sind die jährlichen Totalwerte pro Kanton aus der Pivot-Tabelle. Die relative Veränderung wurde als prozentualer Zuwachs zwischen 2013 und 2023 ausgewiesen.
### 5.2.1 Wachstumsstarke Kantone
Besonders deutlich fällt das Wachstum in den Kantonen Freiburg (+14.8 %), Aargau (+14.2 %), Thurgau (+13.4 %), Waadt (+12.9 %) und Zürich (+12.6 %) aus.

### 5.2.1 Wachsstumsschwache Kantone
Am unteren Ende der Skala finden sich Neuenburg (+1.1 %), Tessin (+3.2 %), Jura (+3.9 %), Graubünden (+5.1 %) sowie Appenzell Innerrhoden (+5.1 %). Diese Kantone verzeichnen entweder eine stagnierende Entwicklung oder nur ein sehr moderates Wachstum. 

```python
# Wachstum berechnen: (2023 / 2013 - 1) * 100
wachstum_prozent = ((pivot_df.loc[2023] / pivot_df.loc[2013]) - 1) * 100
wachstum_prozent = wachstum_prozent.round(2).sort_values()
```

```python
# Balkendiagramm: sortiert nach Wachstum
plt.figure(figsize=(10, 8))
bars = plt.barh(wachstum_prozent.index, wachstum_prozent.values, color="gray")

# Titel und Achsenbeschriftung
plt.title("Prozentuales Bevölkerungswachstum 2013–2023 nach Kanton", fontsize=14)
plt.xlabel("Wachstum in %", fontsize=12)
plt.grid(True, axis='x', linestyle='--', linewidth=0.5)

# Prozentwerte ans Ende jedes Balkens schreiben
for bar in bars:
    breite = bar.get_width()
    plt.text(
        breite + 0.3,                     
        bar.get_y() + bar.get_height() / 2,  
        f"{breite:.1f}%",                 
        va='center',
        fontsize=9
    )

plt.tight_layout()
plt.show()
```

```python
# Top-5 Kantone mit dem höchsten Wachstum
print("Top 5 Kantone mit dem stärksten Wachstum:")
print(wachstum_prozent.tail(5))

# Schwächste 5 Kantone
print("\nSchwächste 5 Kantone:")
print(wachstum_prozent.head(5))
```

## 5.4 Welche Trends lassen sich aus der bisherigen Entwicklung ableiten, und wie könnte sich die Bevölkerung bis 2027 weiterentwickeln?
Um erste Trends zur zukünftigen Entwicklung der Bevölkerung aufzuzeigen, wurde auf Basis der historischen Zeitreihe (2013–2023) ein einfaches Regressionsmodell zur Prognose der Bevölkerungszahlen bis ins Jahr 2027 erstellt. Die Prognose basiert auf einer linearen Regressionsmethode, bei der das jeweilige Kalenderjahr als unabhängige Variable dient und die Bevölkerungszahl als Zielgrösse modelliert wird.

Grundlage der Prognose sind die fünf Kantone mit dem stärksten Bevölkerungswachstum im Zeitraum 2013–2023: Waadt, Thurgau, Aargau, Zürich und Freiburg. Diese Auswahl basiert auf der vorhergehenden Analyse. Für diese Regionen wurde das Modell separat angewendet, um individuelle Wachstumstendenzen zu berücksichtigen.

Die Prognose deutet darauf hin, dass sich der Wachstumstrend in allen fünf Kantonen weiter fortsetzt – mit leichten Unterschieden in der Steigung. Die Modelle zeigen, dass insbesondere Zürich und der Aargau weiterhin ein dynamisches Wachstum aufweisen dürften. Die resultierenden Werte stellen jedoch lediglich eine lineare Fortschreibung der vergangenen Entwicklung dar und bilden externe Einflussfaktoren wie Migration, Geburtenrate oder politische Massnahmen nicht explizit ab.

Die Prognoseergebnisse dienen somit in erster Linie als indikative Szenarien, um zu erkennen, in welchen Regionen mit einer weiter zunehmenden Bevölkerung zu rechnen ist. Für detaillierte Planungen wären komplexere Modelle unter Einbezug zusätzlicher Variablen erforderlich.


```python
# Prognose nur von 2013 bis 2027 (inkl. 4 Jahre Vorausblick)
prognose_df = pd.DataFrame(index=range(2013, 2028), columns=pivot_df.columns)

for kanton in pivot_df.columns:
    y = pivot_df[kanton].dropna()
    x = np.array(y.index)
    y_values = y.values

    slope, intercept, _, _, _ = linregress(x, y_values)

    x_future = np.arange(2013, 2028)
    y_pred = intercept + slope * x_future

    prognose_df[kanton] = y_pred.round(0)
```

```python
plt.figure(figsize=(10, 6))

# Schattierung für Prognosezeitraum (optional, aber hilfreich)
plt.axvspan(2023.5, 2027.5, color="gray", alpha=0.1, label="Prognosezeitraum")

# Prognosekurven für Top 5 Kantone
for kanton in top5_kantone:
    plt.plot(
        prognose_df.index,
        prognose_df[kanton],
        label=kanton,
        marker="o",
        linewidth=2
    )

    # Endwert für 2027 als Text anzeigen
    endwert = prognose_df[kanton].loc[2027]
    plt.text(
        x=2027 + 0.2,
        y=endwert,
        s=f"{int(endwert):,}".replace(",", "'"),
        va="center",
        fontsize=9
    )

# Formatierung
plt.title("Prognose 2024–2027 – Kantone mit dem höchsten Wachstum", fontsize=14)
plt.ylabel("Einwohnerzahl", fontsize=12)
plt.xlabel("Jahr", fontsize=12)
plt.grid(True, linestyle="--", linewidth=0.5)
plt.xticks(prognose_df.index)
plt.legend(title="Kanton", loc="upper left")
plt.tight_layout()
plt.show()
```

# 6. Fazit und Ausblick

Die Analyse der Schweizer Bevölkerungsentwicklung zwischen 2013 und 2023 zeigt ein stabiles und kontinuierliches Wachstum auf nationaler Ebene von rund 9 %. Dieses Wachstum verlief ohne spürbare Rückgänge oder Unterbrechungen, was auf eine resiliente demografische Entwicklung schliessen lässt.

Auf kantonaler Ebene treten jedoch markante Unterschiede zutage: Während Kantone wie Freiburg, Aargau, Thurgau, Waadt und Zürich teils deutlich über dem nationalen Schnitt gewachsen sind, blieb die Entwicklung in Neuenburg, Tessin, Jura oder Appenzell Innerrhoden deutlich verhaltener. Diese Differenzen deuten auf strukturelle Faktoren hin – etwa die Attraktivität urbaner Zentren hinsichtlich Arbeitsplätzen, Infrastruktur und Bildungseinrichtungen im Vergleich zu ländlich geprägten Regionen mit geringerer Zuwanderung und begrenzten Wachstumsimpulsen.

Die erstellte Prognose bis ins Jahr 2027 zeigt, dass sich diese Trends voraussichtlich fortsetzen. Allerdings ist zu betonen, dass das verwendete Modell auf einer einfachen linearen Regression basiert. Es berücksichtigt keine externen Einflussfaktoren wie Geburtenraten, Sterblichkeit, Migration oder wirtschaftliche Rahmenbedingungen. Auch politische Massnahmen wie Wohnungspolitik, Zuwanderungsregulierung oder steuerliche Anreize bleiben unberücksichtigt.

Diese Limitationen machen deutlich, dass die vorliegenden Ergebnisse als indikative Trends zu interpretieren sind – nicht als belastbare Prognosen im engeren Sinne. Für eine fundiertere Zukunftsmodellierung wäre der Einbezug zusätzlicher Variablen und komplexerer Modelle notwendig.

Nicht zuletzt unterstreicht die Analyse den wachsenden Handlungsdruck in stark wachsenden Kantonen: Raumplanung, Verkehrsinfrastruktur, Bildung und Gesundheitswesen müssen vorausschauend weiterentwickelt werden, während strukturschwächere Regionen gezielte Impulse benötigen, um einer zunehmenden demografischen Polarisierung entgegenzuwirken.
